```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualisasi
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_taniland_v2.csv')

df.head()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.info
```

```python
sns.countplot(x=df['Kelembaban Tanah'])
```

```python
plt.figure(1 , figsize = (15 , 6))
n = 0 
for x in ['Kelembaban Tanah' , 'Kelembaban Udara' , 'Suhu']:
    n += 1
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.distplot(df[x] , bins = 15)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
sns.pairplot(df[[ 'Kelembaban Tanah', 'Kelembaban Udara', 'Suhu']])
```

Plotting data berdasarkan Musim

```python
sns.pairplot(df, vars = ['Kelembaban Tanah', 'Kelembaban Udara', 'Suhu'], hue = "Tanaman")
```

Plotting data berdasarkan Iklim

```python
sns.pairplot(df, vars = ['Kelembaban Tanah', 'Kelembaban Udara', 'Suhu'], hue = "Iklim")
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
plt.rcParams['figure.figsize'] = (14, 8)
sns.heatmap(df[[ 'Kelembaban Tanah', 'Kelembaban Udara', 'Suhu']].corr(), cmap = 'magma_r', annot = True, linewidths=.5)
plt.title('Heatmap', fontsize = 20)
plt.show()
```

```python
plt.figure(1 , figsize = (15 , 7))
plt.title('Scatter plot of Age v/s Spending Score', fontsize = 20)
plt.xlabel('Kelembaban Tanah')
plt.ylabel('Suhu')
plt.scatter( x = 'Kelembaban Tanah', y = 'Suhu', data = df, s = 100)
plt.show()
```
